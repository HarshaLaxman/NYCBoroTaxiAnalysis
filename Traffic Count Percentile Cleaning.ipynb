```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import patsy
import statsmodels.api as sm
from scipy.stats import ttest_ind
```

```python
df1 = 'Cleaned_Traffic_Count.csv'
df1 = pd.read_csv(df1)

```

```python
del df1['Unnamed: 0']
```

```python
df1.head(n=5)
```

```python
#df_change will be the hourly increase/decrease of traffic in % each day.The first hour will be based on the 
#previous day's traffic if it exists.
df2 = df1.copy()
df2
```

```python
del df2['Roadway Name']
del df2['From']
del df2['To']
del df2['ID']
```

```python
len(df1)
```

```python
#find the percentile change in traffic between every hour
#loop through each row and the before and after midnight times
#perform the calculation and save to df2 for non-zero values, changes of traffic from and value to 0 cant be determined

for i in range(len(df1)):
    for j in range(5,28):
        x = df1.iloc[i,j]
        y = df1.iloc[i,j+1]
        if x>0:
            z = 100*(y-x)/x
            df2.iloc[i,j+1] = z
```

```python
#find the percentile change in traffic during midnight between 2 days
#loop through each row and the before and after midnight times
#if the 2 times are of the same segment, then perform the calculation and save to df2

for i in range(len(df1)-1):
    if df1.iloc[i,2] == df1.iloc[i+1,2]:
        if df1.iloc[i,3] == df1.iloc[i+1,3]:
            x = df1.iloc[i,28]
            y = df1.iloc[i+1,5]
            if x>0:
                z = 100*(y-x)/x
                df2.iloc[i+1,5] = z
```
